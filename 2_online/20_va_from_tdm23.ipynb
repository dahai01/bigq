# VA from TDM23

```python
import matplotlib.pyplot as plt
import pandas as pd 
import sqlite3
```

## Inputs

```python
tdm_db = r"D:\apps\myproject\tdm23\outputs\Base\tdm23.db"
```

## DB data

```python
con = sqlite3.connect(tdm_db) 

query_string = '''SELECT M.mpo, V.veh_suff as va, COUNT(H.hid) as num_households
FROM hh H
JOIN veh V ON H.hid = V.hid
JOIN MA_taz_geography M ON H.taz_id = M.taz_id
WHERE H.drivers <= 3
GROUP BY M.mpo, V.veh_suff;
    '''
query_df = pd.read_sql(query_string, con) 

con.close() 
```
